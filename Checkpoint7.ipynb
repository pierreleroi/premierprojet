```python
import numpy as np
import pandas as pd

exam = ({
    'Noms':('Anastasia', ('Dima'), ('Katherine'), ('James'), ('Emily'), ('Michael'), ('Matthieu'), ('Laura'), ('Kevin'), ('Jonas')),
    'Scores' : (12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19),
    'Tentatives' : (1, 3, 2, 3, 2, 3, 1, 1, 2, 1),
    'Qualifier' : (['oui', 'non', 'oui', 'non', 'non', 'oui', 'oui', 'non', 'non', 'oui'])
})

etiq = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(exam, index = etiq)
print("\n---:le score de chaque joueur:---")
print("\n",df)
df1 = df.head(3)
print("\n---: 1.voici les 3 premières lignes du dictionnaire:---\n")
print(df1)
print("\n---: 2.Suppression des lignes avec des valeurs NaN : ---")
print(df.dropna())
print("\n---: 3. Extraire les colonnes 'Noms' et 'Scores' : ---")
print("\n",df[["Noms","Scores"]])
#indexis = ['k']
#nv_lig = pd.DataFrame([["Sureh", 15.5, 1, "Oui"]], columns = df.columns, index = indexis)
#print("\n",nv_lig)
#df = df.append(nv_lig, ignore_index=True, sort=True)
# la methode .loc() nous permet d'ajouter une ligne avec son index dans le tableau
df.loc['k'] = ["Sureh", 15.5, 1, "Oui"]
print("\n---:4. La Nouvelle ligne k à ajouter est :")
print("\n",df)
print("\n---: 5. la colonne 'Tentatives' est supprimée : ---")
print("\n",df.drop(["Tentatives"], axis = 1))
print("\n------6 .Ajout d'une nouvelle colonne succès, si le score est spérieur à 10---on a 1 ---sinon---on a 0 \n")
df ['Succès'] = np.where(df['Scores'] > 10, '1', '0')
#df['Succès'] = ['bon' if s > 10 else "mauvais" for s in df['Scores'] ]
print(df)
```
